```python
# Handled the UNSPECIFIED countries
# Since I do not want to blindly remove the UNSPECIFIED countries rows to avoid losing real customer history
# I checked to see if customers with UNSPECIFIED countries have valid countries elsewhere
# If they do, I replace the UNSPECIFIED countries with their valid countries
# Otherwise, I remove the UNSPECIFIED countries


# I initially wrote the following code to handle the UNSPECIFIED countries
# customers_with_known_countries = product_sales_data.loc[product_sales_data['Country'] != 'UNSPECIFIED', 'Customer ID'].unique()
# product_sales_data = (product_sales_data[(product_sales_data['Country'] != 'UNSPECIFIED') | (product_sales_data['Customer ID']
#                                                                                              .isin(customers_with_known_countries))])


# But I decided to settle for the following code because it is smarter
# I used the following code to create a dictionary with each unqiue Customer ID mapped to it's corresponding Country
customer_country_connection = (product_sales_data.loc[product_sales_data['Country'] != 'UNSPECIFIED', ['Customer ID', 'Country']]
                               .drop_duplicates(subset = 'Customer ID')
                               .set_index('Customer ID')['Country']
                               .to_dict())


# I used the following to rewrite the UNSPECIFIED countries of customers with known countries elsewhere
product_sales_data['Country'] = product_sales_data.apply(
    lambda row: customer_country_connection.get(row['Customer ID'], row['Country']),
    axis = 1
)


# After the replacement, I removed any UNSPECIFIED countries that are left
product_sales_data = product_sales_data[product_sales_data['Country'] != 'UNSPECIFIED']
```

```python
# Made a copy of the product_sales_data
# Started building the customer_level_data

sales_data = product_sales_data.copy()

sales_data['InvoiceDate'] = pd.to_datetime(sales_data['InvoiceDate'])
sales_data['Revenue'] = sales_data['Quantity'] * sales_data['Price']
sales_data['PurchaseQty'] = sales_data['Quantity'].clip(lower = 0)
sales_data['ReturnQty'] = (-sales_data['Quantity']).clip(lower = 0)

sales_data['PurchaseRevenue'] = sales_data['PurchaseQty'] * sales_data['Price']
sales_data['ReturnRevenue'] = sales_data['ReturnQty'] * sales_data['Price']

sales_data
```

```python
customer_level_data = sales_data.groupby('Customer ID').agg(PurchaseQty = ('PurchaseQty', 'sum'),
                                                            ReturnQty = ('ReturnQty', 'sum'),
                                                            PurchaseRevenue = ('PurchaseRevenue', 'sum'),
                                                            ReturnRevenue = ('ReturnRevenue', 'sum')).reset_index()


customer_level_data['NetRevenue'] = customer_level_data['PurchaseRevenue'] - customer_level_data['ReturnRevenue']

customer_level_data[['PurchaseQty', 'ReturnQty', 'PurchaseRevenue', 'ReturnRevenue']] = \
  customer_level_data[['PurchaseQty', 'ReturnQty', 'PurchaseRevenue', 'ReturnRevenue']].fillna(0)

customer_level_data
```

```python
# Calculated the NumPurchases and NumReturns (Purchase Frequency and Return Frequency) features

purchase_frequency = (sales_data.loc[~sales_data['Invoice'].str.startswith('C', na = False)]
                      .groupby('Customer ID')['Invoice']
                      .nunique()
                      .reset_index(name = 'NumPurchases'))


return_frequency = (sales_data.loc[sales_data['Invoice'].str.startswith('C', na = False)]
                    .groupby('Customer ID')['Invoice']
                    .nunique()
                    .reset_index(name = 'NumReturns'))

customer_level_data = customer_level_data.merge(purchase_frequency, on = 'Customer ID', how = 'left')
customer_level_data = customer_level_data.merge(return_frequency, on = 'Customer ID', how = 'left')

customer_level_data[['NumPurchases', 'NumReturns']] = customer_level_data[['NumPurchases', 'NumReturns']].fillna(0).astype(int)

customer_level_data
```

```python
# Started calculating the RecencyDays and TenureDays

reference_date = sales_data['InvoiceDate'].max() + pd.Timedelta(days = 1)
reference_date
```

```python
# Calculated the RecencyDays and TenureDays from FirstPurchaseDate and LastPurchaseDate

purchase_transactions = sales_data.loc[~sales_data['Invoice'].str.startswith('C', na = False)]

customer_timespent = purchase_transactions.groupby('Customer ID').agg(FirstPurchaseDate = ('InvoiceDate', 'min'),
                                                       LastPurchaseDate = ('InvoiceDate', 'max')).reset_index()

customer_timespent['RecencyDays'] = (reference_date - customer_timespent['LastPurchaseDate']).dt.days
customer_timespent['TenureDays'] = (customer_timespent['LastPurchaseDate'] - customer_timespent['FirstPurchaseDate']).dt.days

customer_timespent
```

```python
# Continue here when I come back
customer_level_data = customer_level_data.merge(customer_timespent, on = 'Customer ID', how = 'left')

# I didn't fill NaN values in the RecencyDays with 0, since customers who do not have any purchase history do not also have any RecencyDays
# So i dropped them
customer_level_data = customer_level_data.dropna(subset = ['RecencyDays'])
customer_level_data[['RecencyDays', 'TenureDays']] = (customer_level_data[['RecencyDays', 'TenureDays']]).astype(int)
customer_level_data
```

```python
# Calculated the AvgDaysBetweenPurchases feature

avgdays_transactions = sales_data.loc[~sales_data['Invoice'].str.startswith('C', na = False)].copy()
avgdays_transactions['InvoiceDate'] = pd.to_datetime(avgdays_transactions['InvoiceDate'])

# I sorted the values per Customer ID and InvoiceDate so that I can compute the days differences per transaction and get the AvgDaysBetweenPurchases
avgdays_transactions = avgdays_transactions.sort_values(['Customer ID', 'InvoiceDate'])

# Calculated the days difference per transaction
avgdays_transactions['DaysBetweenPurchases'] = (avgdays_transactions.groupby('Customer ID')['InvoiceDate']
                                               .diff().dt.days)

# Calculated the mean value which represents the AvgDaysBetweenPurchase
avgdays = (avgdays_transactions.groupby('Customer ID')['DaysBetweenPurchases']
           .mean().reset_index(name = 'AvgDaysBetweenPurchases'))

# Merged the avgdays dataframe into the customer_level_data dataframe
customer_level_data = customer_level_data.merge(avgdays, on = 'Customer ID', how = 'left')

# Handled AvgDaysBetweenPurchases for one-time buyers
customer_level_data['AvgDaysBetweenPurchases'] = (customer_level_data['AvgDaysBetweenPurchases'].fillna(customer_level_data['TenureDays']))

# Checked the customer_level_data to see what it looked like
customer_level_data
```

```python
# Calculated the ReturnRatio, AvgBasketSize, and AvgOrderValue features

customer_level_data['ReturnRatio'] = customer_level_data['ReturnQty'] / customer_level_data['PurchaseQty']
customer_level_data['ReturnRatio'] = customer_level_data['ReturnRatio'].replace([np.inf, -np.inf], np.nan).fillna(0)

customer_level_data['AvgBasketSize'] = customer_level_data['PurchaseQty'] / customer_level_data['NumPurchases']
customer_level_data['AvgBasketSize'] = customer_level_data['AvgBasketSize'].replace([np.inf, -np.inf], np.nan).fillna(0)

customer_level_data['AvgOrderValue'] = customer_level_data['PurchaseRevenue'] / customer_level_data['NumPurchases']
customer_level_data['AvgOrderValue'] = customer_level_data['AvgOrderValue'].replace([np.inf, -np.inf], np.nan).fillna(0)

customer_level_data
```

```python
# Calculated the dormancy feature
tolerance_factor = 3

customer_level_data['IsDormant'] = (customer_level_data['RecencyDays']
                                    > tolerance_factor * customer_level_data['AvgDaysBetweenPurchases']).astype(int)

customer_level_data
```
